# Analyze



Create new column DayOfTheWeek by generating date in the form of day of the week for further analysis.

```r
daily_activity <- daily_activity%>%
    mutate(DayOfWeek = weekdays(as.Date(daily_activity$ActivityDate)))
```

Constructing totel_mins and totel_hours 

```r
daily_activity <- daily_activity%>%
    mutate(totel_mins = as.numeric(daily_activity$VeryActiveMinutes+daily_activity$FairlyActiveMinutes+daily_activity$LightlyActiveMinutes+daily_activity$SedentaryMinutes))
```

```r
daily_activity <- daily_activity%>%
    mutate(totel_hours = as.numeric(daily_activity$totel_mins)/60)
```

```r
head(daily_activity)
```

1. Based on this above table, we can see that on average, users walked 7638 steps. CDC recommended that 10,000 steps per day is a suitable goal for an adult to have a better heatlh, weigh loss and fitness improvement. (Source: How many steps should people take per day?)

2. Most of users logged in devices is Sedentary user, which means they may use device for normal activities, not focus on fitness only. We need to find out more later.

Adding new column "Day"- Day of week to "sleep_day" data

Create new column DayOfTheWeek by generating date in the form of day of the week for further analysis.

```r
daily_sleep <- daily_sleep%>%
    mutate(Day = weekdays(as.Date(daily_sleep$SleepDay)))
```

```r
daily_sleep <- daily_sleep %>%
    mutate(TotalTimeAwakeOnBed = as.numeric(daily_sleep$TotalTimeInBed - daily_sleep$TotalMinutesAsleep))

head(daily_sleep)
```

Let's create a new column name "hour" and calculate the average step per hour in hourly_steps table

```r
hourly_steps <- hourly_steps%>%
    mutate(time = as.numeric(format(hourly_steps$ActivityHour, format = "%H")))

hourly_steps <- hourly_steps%>%
                #group_by(hourly_steps$time)%>%
                mutate(average_steps = (hourly_steps$StepTotal) /mean(hourly_steps$StepTotal ))
head(hourly_steps)
```

# Share

Now, we should creating visualizations of data and communicating our findings based on our analysis with our stakeholder

**Plotting a chart to show the percentage of activity in Mins**

```r
# This function help to resize the plots
fig <- function(width, heigth){options(repr.plot.width = width, repr.plot.height = heigth)}
fig(16,8)
```

```r
# calculating the total mins for each column
very_active_mins <- sum(daily_activity$VeryActiveMinutes)
fairly_active_mins <- sum(daily_activity$FairlyActiveMinutes)
lightly_active_mins <- sum(daily_activity$LightlyActiveMinutes)
sedentary_mins <- sum(daily_activity$SedentaryMinutes)

y <- very_active_mins+fairly_active_mins+lightly_active_mins+sedentary_mins
z <- 100/y

# plotting pie chart
x <- c(very_active_mins*z , fairly_active_mins*z , lightly_active_mins*z , sedentary_mins*z)
labels <- c("very_active_mins","fairly_active_mins","lightly_active_mins","sedentary_mins")

color <- brewer.pal(length(labels),"Set2")
pie(x,labels = paste0(round(x,2), "%"), col = color , radius = 1)

legend("topright",legend = labels , fill = color)
```

From the pie chart:

* Sedentary minutes takes the biggest slice (81.3%). This indicates that users are using the FitBit app to log daily activities such as daily commute, inactive movements, etc...

* There are small percentage of user ussing Firbit to track thier fitness activity (1.1% of Fairly active minutes, and 1.7% of Very active minute)

```r
Frequency <- daily_activity%>%count(DayOfWeek)
DayOfWeek <-c(Frequency$DayOfWeek)
n <-c(Frequency$n)
barplot(n,names.arg=DayOfWeek,main="No. of times users logged in app in one week",
       xlab="Day of the week",ylab="Frequency",ylim=)
```

By looking in this bar chart, we can see the frequency of user logged in FitBit app for each day of the week and user more using it on the weekday than weekend (from Tuesday to Friday and then a bit decreased after that).


Lets see Average steps taken for every hour

```r
SUM <- hourly_steps%>%
    group_by(time)%>%
    summarise(Step_taken = (sum(average_steps)))

ggplot(SUM)+
    geom_bar(aes(x=time , y=Step_taken), stat = 'identity')+
    ggtitle("Average Steps Taken for Every Hour")
```

A large numebr of step taken happenend from 08:00 to 20:00, which means on user daily basis. plus wwith the chart of No. times user logged in app in one week, we can concluded that they prefer using Firbit on Working day.

The highest steps taken are around 18:00, which should be the hour when most users end their job and start to do their exercise routines.

lets see is there a relationship between Steps Taken and Calories Burned

```r
ggplot(daily_activity,aes(x=TotalSteps	, y=Calories,color=Calories,size=10 , alpha = 0.7))+
geom_point()+
ggtitle("Calories burned for every step taken")

```
